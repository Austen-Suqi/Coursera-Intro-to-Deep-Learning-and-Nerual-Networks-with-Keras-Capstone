### Load all necessary libraries

```python
import pandas as pd
import numpy as np
```

### Load data

```python
concrete_data = pd.read_csv('https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DL0101EN/labs/data/concrete_data.csv')
concrete_data.head()
```

From the lab we know there are 1,030 data points and there are no blank values

### Split data into dependent and independent variables (X and y)

```python
concrete_data_columns = concrete_data.columns

X = concrete_data[concrete_data_columns[concrete_data_columns != 'Strength']] # all columns except Strength
y = concrete_data['Strength'] # Strength column

n_cols= X.shape[1]
```

```python
y.head()
```

### Load Keras

```python
!pip install TensorFlow
!pip install keras
import tensorflow
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

### Build the model for part A

```python
# define regression model
def regression_model():
    # create model
    model = Sequential()
    model.add(Dense(10, activation='relu', input_shape=(n_cols,)))
    model.add(Dense(1))
    
    # compile model
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model
```

### Split the data

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
```

### Create and fit the model

```python
modelA=regression_model()
modelA.fit(X_train,y_train,validation_split=0.3,epochs=50)
yhat=modelA.predict(X_test)
```

### Find the mean squared error

```python
from sklearn.metrics import mean_squared_error
MSE=np.array([mean_squared_error(y_test,yhat)])
print('The mean squared error is: %d' %MSE)
```

```python
yhat[0:5]
```

```python
y_test[0:5]
```

### For loop to create 50 MSEs

```python
for x in range (1,50):
    modelA.fit(X_train,y_train,validation_split=0.3,epochs=50)
    yhat=modelA.predict(X_test)
    MSE =np.vstack([MSE,mean_squared_error(y_test,yhat)])
```

```python
print('The mean of the mean squared error is: %d' %MSE.mean())
print('The standard deviation of the mean squared error is: %d' %MSE.std())
```

### Part B - Normalized Data

```python
X_norm=(X-X.mean())/X.std()
X_norm.head()
```

### Split the normalized data

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_norm, y, test_size=0.30, random_state=42)
```

### Fit the model (using the same one built as part A, but fitting with normalized data)

```python
modelB=regression_model()
modelB.fit(X_train,y_train,validation_split=0.3,epochs=50)
yhatB=modelB.predict(X_test)
```

### Find the MSE

```python
MSEB=np.array([mean_squared_error(y_test,yhatB)])
print('The mean squared error is: %d' %MSEB)
```

### For loop to create 50 MSEs

```python
for x in range (1,50):
    modelB.fit(X_train,y_train,validation_split=0.3,epochs=50)
    yhatB=modelB.predict(X_test)
    MSEB =np.vstack([MSEB,mean_squared_error(y_test,yhatB)])
```

### The mean of the mean squared error is smaller than that of part A, but the standard deviation is higher

```python
print('The mean of the mean squared error is: %d' %MSEB.mean())
print('The standard deviation of the mean squared error is: %d' %MSEB.std())
```

### Part D - 3 hidden layers

### Build the model for part D

```python
# define regression model
def regression_model():
    # create model
    model = Sequential()
    model.add(Dense(10, activation='relu', input_shape=(n_cols,)))
    model.add(Dense(10, activation='relu'))
    model.add(Dense(10, activation='relu'))
    model.add(Dense(1))
    
    # compile model
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model
```

### Fit the model (using the same one built as part B, but with 3 hidden layers as built above)

```python
modelD=regression_model()
modelD.fit(X_train,y_train,validation_split=0.3,epochs=50)
yhatD=modelD.predict(X_test)
```

### Find the MSE

```python
MSED=np.array([mean_squared_error(y_test,yhatD)])
print('The mean squared error is: %d' %MSED)
```

### For loop to create 50 MSEs

```python
for x in range (1,50):
    modelD.fit(X_train,y_train,validation_split=0.3,epochs=50)
    yhatD=modelD.predict(X_test)
    MSED =np.vstack([MSED,mean_squared_error(y_test,yhatD)])
```

### The mean of the mean squared error is smaller than that of part B, and the standard deviation is also lower

```python
print('The mean of the mean squared error is: %d' %MSED.mean())
print('The standard deviation of the mean squared error is: %d' %MSED.std())
```
